```python
file_in="/home/allan/Desktop/new_haven_seeclickfix_visualization/data/new_haven.xml"
data = ET.parse(open(file_in))
```

```python
names={}
for elem in data.findall("./way/tag[@k='name']"):
    street_name = elem.attrib["v"]
    if street_name in names.keys():
        names[street_name]+=1
    else:
        names[street_name]=1
sorted(names.items(), key=lambda x: x[1],reverse=True)
```

```python
refs=[]
for i in data.findall("./way/tag[@k='name'][@v='Dwight Street'].."):
    ref=[]
    for j in i.findall("./nd"):
        ref.append(j.attrib['ref'])
    refs.append(ref)
refs
plt.figure()
for ref in refs:
    for ndid in ref:
        node = data.findall("./node[@id='"+ndid+"']")[0]
        plt.plot(node.attrib["lon"],node.attrib["lat"],'rx')
refs=[]
for i in data.findall("./way/tag[@k='name'][@v='Chapel Street'].."):
    ref=[]
    for j in i.findall("./nd"):
        ref.append(j.attrib['ref'])
    refs.append(ref)
refs
for ref in refs:
    for ndid in ref:
        node = data.findall("./node[@id='"+ndid+"']")[0]
        plt.plot(node.attrib["lon"],node.attrib["lat"],'bx')
```
